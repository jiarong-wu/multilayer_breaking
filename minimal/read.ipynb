### Read surface elevation

```python
case = Case(NL=15, LEVEL=10, path='/projects/DEIKE/jiarongw/multilayer/revision/field_new_200m_P0.02_RE40000_10_15_rand2_Htheta0.503/')
case.label = 'C4_rand2'
tseries = (0,40,80,120,160,110,130,150,170)

newdir = '/projects/DEIKE/jiarongw/multilayer/JPO/' + case.label
os.mkdir(newdir)


for t in tseries: 
    print('Processing t=%g' %t)
    
    # eta
    filename = case.path + 'surface/'+ 'eta' + '_matrix_%g' %t
    matrix = np.fromfile(filename, dtype=np.float32)
    N = 2**case.LEVEL
    
    oldfilename = '../data_dissipation/compilation/fieldsample_' + case.label+'_t%g' %t + '.nc'
    newfilename = newdir + '/field%g' %t + '.nc'
    ds = xr.open_dataset(oldfilename)
    ds['eta'] = (('x','y'), matrix)
    encoding = {}
    for var_name in ds.data_vars:
        encoding[var_name] = {'dtype': 'float32', 'zlib': True}
    ds.to_netcdf(newfilename, mode='w', encoding=encoding, format='NETCDF4')
```
